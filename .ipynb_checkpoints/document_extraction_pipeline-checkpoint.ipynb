````python
# Analyze mathematics data
theorem_counts = Counter([r.get('theorem_number') for r in mathematics_data if r.get('theorem_number')])

print("\nğŸ“Š MATHEMATICS EXTRACTION SUMMARY")
print("â•" * 65)
print(f"\nğŸ“š Total Records: {len(mathematics_data)}")
print(f"\nğŸ”¢ Theorem Types:")
for theorem, count in theorem_counts.most_common():
    print(f"   â€¢ {theorem}: {count} occurrences")

print("\nğŸ“– Main Topics:")
print("   â€¢ Euler's Theorem applications")
print("   â€¢ Differential Calculus concepts")
print("   â€¢ Partial derivatives")
print("\nâœ… Primary Focus: Unit II - Differential Calculus-I")
````

## ğŸ¤– SLM Extraction Strategy

<div style="background: linear-gradient(135deg, #f3e5f5 0%, #e1bee7 100%); padding: 20px; border-radius: 10px; border-left: 5px solid #9c27b0;">
    <h3 style="color: #6a1b9a; margin-top: 0;">ğŸ§  Ollama Phi3 Configuration</h3>
    <p><strong>Model:</strong> Phi3 (3.8B parameters)</p>
    <p><strong>Temperature:</strong> 0.1 (low for consistent extraction)</p>
    <p><strong>Output:</strong> JSON-structured with domain-specific fields</p>
    <p><strong>Fallback:</strong> Regex-based extraction if JSON parsing fails</p>
</div>

````python
# Sample extraction prompt
print("\nğŸ¯ SAMPLE EXTRACTION PROMPT (Constitution)")
print("â•" * 65)

print("\nSYSTEM PROMPT:")
print("â”€" * 65)
print("""You are an expert at extracting structured information from Indian 
Constitution documents. Your task is to identify articles, parts, 
and chapters with high precision. Always output valid JSON with 
exact field names as specified.""")

print("\nUSER PROMPT TEMPLATE:")
print("â”€" * 65)
print("""Extract information from this Indian Constitution text.

TEXT:
{text}

INSTRUCTIONS:
1. Find the Article number (e.g., "Article 14", "Article 21")
2. Extract the article title/heading
3. Identify the PART (e.g., "PART III", "PART I")
4. Identify the CHAPTER if mentioned
5. If information is not present, use null

OUTPUT FORMAT (valid JSON only):
{
    "article_number": "Article X or null",
    "article_title": "Title text or null",
    "part": "PART X or null",
    "chapter": "CHAPTER X or null"
}

Return ONLY the JSON object, no other text.""")

print("\nEXPECTED OUTPUT:")
print("â”€" * 65)
print(json.dumps({
    "article_number": "Article 14",
    "article_title": "Equality before law",
    "part": "PART III",
    "chapter": "CHAPTER I"
}, indent=2))
````
